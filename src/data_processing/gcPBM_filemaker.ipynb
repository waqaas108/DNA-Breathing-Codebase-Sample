```python
import numpy as np
import pandas as pd
import pickle
import os
```

```python
# load the dictionaries from pickle files
with open('../../data/gcPBM_data/mad.pkl', 'rb') as handle:
    mad_data = pickle.load(handle)

with open('../../data/gcPBM_data/max.pkl', 'rb') as handle:
    max_data = pickle.load(handle)

with open('../../data/gcPBM_data/myc.pkl', 'rb') as handle:
    myc_data = pickle.load(handle)
```

```python
mad_df = {}
for partition in mad_data.keys():
    part = mad_data[partition]
    for seq, val in part.items():
        mad_df[seq] = val['seq']

max_df = {}
for partition in max_data.keys():
    part = max_data[partition]
    for seq, val in part.items():
        max_df[seq] = val['seq']

myc_df = {}
for partition in myc_data.keys():
    part = myc_data[partition]
    for seq, val in part.items():
        myc_df[seq] = val['seq']
```

```python
with open ('../../data/gcPBM_data/mad_gcPBM.txt', 'w') as handle:
    for key, value in mad_df.items():
        handle.write(str(key) + ' ' + value + '\n')

with open ('../../data/gcPBM_data/max_gcPBM.txt', 'w') as handle:
    for key, value in max_df.items():
        handle.write(str(key) + ' ' + value + '\n')

with open ('../../data/gcPBM_data/myc_gcPBM.txt', 'w') as handle:
    for key, value in myc_df.items():
        handle.write(str(key) + ' ' + value + '\n')
```

```python
pkl_dir = '/project2/xinhe/waqaas/DNA-breathing/data/gcPBM_features/bubble_data/mad_gcPBM'
```

```python
new_data = {}
for file in os.listdir(pkl_dir):
    if file.endswith('.pkl'):
        with open(os.path.join(pkl_dir, file), 'rb') as handle:
            data = pickle.load(handle)
            new_data[file.split('.')[0]] = data
```

```python
mad_df = pd.DataFrame(new_data)
```
