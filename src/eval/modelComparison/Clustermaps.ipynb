```python
import torch
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.cluster import hierarchy
```

```python
df = pd.read_csv('Modelmatrix.csv', header=None)
```

```python
models = ['No normalisation, nullification at inference', 'Normalised features, compared with Gaussian noise', 'Normalised features, Gaussian noise at inference', 'Normalised features, nullification at inference']
```

```python
linkage = hierarchy.linkage(df.values, method='ward')
labels = [i for i in models]
clustermap_all = sns.clustermap(df, row_linkage=linkage, col_linkage=linkage, xticklabels=labels, yticklabels=labels, figsize = (10, 10), cmap="mako_r", annot=True, vmin=0, vmax=20)
clustermap_all.savefig("Modelcomparison.png")
```
