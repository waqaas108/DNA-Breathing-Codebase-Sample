```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the barplot
left_bar_heights = [0.9758, 0.9615, 0.9434]
right_bar_heights = [0.9755, 0.9610, 0.9429]
bar_width = 0.4

# X-axis positions for the bars
x = np.arange(len(left_bar_heights))

# Create the barplot
fig, ax = plt.subplots()
ax.bar(x - bar_width/2, left_bar_heights, bar_width, color='blue', label='Sequence Only')
ax.bar(x + bar_width/2, right_bar_heights, bar_width, color='lightblue', label='Sequence + Bubble')

# Set labels and title
ax.set_xlabel('TFs')
ax.set_ylabel('r2')
ax.set_title('r2 of Models for each TF')

# Set x-axis tick labels
group_labels = ['MAD', 'MAX', 'MYC']
ax.set_xticks(x)
ax.set_xticklabels(group_labels)

# Set y-axis limits
ax.set_ylim(0.9, 1.0)

# Add legend
ax.legend()

# Show the plot
plt.show()
```
